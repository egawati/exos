## Varying number of streams

- Number of streams = {5,10,15,20,25,30, 35, 40, 45}
- Total data per stream = 100K (1% outlier)
- Window size = 1000
- Number of principal component = 1
- Case 1 : data attributes are strongly correlated within each stream and across other streams
- Number of attributes in each stream varies : {3, 7, 10, 16}


```python
bname = '100K_Case1'
rel_path = 'dbpca/pickles/performance/nstreams'
df = pd.read_pickle(f'{rel_path}/avg_performance_{bname}.pkl')
df
```

```python
Xs= (df['nstreams'],)
y = (df['avg_precision'],)
xlabel = 'Number of data stream sources'
ylabel = 'Average precision'
lines = ('go--', )
labels = ('EXOS', )
title = 'Generating outlying attributes on Case 1 Dataset'
plot_nstreams_performance(Xs, y, xlabel, ylabel, lines, labels, title, loc='lower right')
```
