```python
import pandas as pd
pd.set_option('display.precision', 2)
from ipynb.fs.full.graphs import plot_categorical_performance
from ipynb.fs.full.graphs import draw_multiple_plots_by_cases
```

## Varying the degree of data attribute correlations

- Degree of correlations: {Case 1, Case 2, Case 3, Case 4}
- Number of streams = 15
- Total data per stream = 100K (1% outlier)
- Window size = 1000
- Number of principal component = 1
- Number of attributes in each stream varies : {3, 7, 10, 16}
- Run each number of streams 10 times



```python
bname = '100K'
rel_path = 'dbpca/pickles/performance/cases'
df_exos_dbpca = pd.read_pickle(f'{rel_path}/avg_performance_{bname}.pkl')
df_exos_dbpca
```

```python
bname = '100K'
rel_path = 'pca/pickles/performance/cases'
df_exos_pca = pd.read_pickle(f'{rel_path}/avg_performance_{bname}.pkl')
df_exos_pca
```

```python
xval = df_exos_dbpca['case']
yvals = {}
yvals['precision'] = (df_exos_dbpca['avg_precision'], df_exos_pca['avg_precision'],)
yvals['recall'] = (df_exos_dbpca['avg_recall'], df_exos_pca['avg_recall'],)
yvals['f1_score'] = (df_exos_dbpca['avg_f1_score'], df_exos_pca['avg_f1_score'],)
nsubplots = 3
figsize = (18,3)
lines = ('go--', 'bs:',)
labels = ('EXOS DBPCA', 'EXOS PCA')
title = 'Measuring performance'
xlabel = 'Data attribute correlation'
ylabels = ('Average precision', 'Average recall', 'Average f1 score')

draw_multiple_plots_by_cases(yvals, xval, nsubplots, figsize, 
                             lines, labels, title, 
                             ylabels, xlabel,
                             loc='lower right', sharey=True)
```

```python
Xs= (df_exos_dbpca['case'],df_exos_pca['case'],)
ys = (df_exos_dbpca['avg_precision'], df_exos_pca['avg_precision'],)
xlabel = 'Data attribute correlation'
ylabel = 'Average precision'
lines = ('go--', 'bs-',)
labels = ('EXOS DBPCA', 'EXOS PCA')
title = ''
plot_categorical_performance(Xs, ys, xlabel, ylabel, lines, labels, title, loc='lower right')
```
