```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
folder = 'MDist'
nstreams = 4
bfname = 'cd_mdist'
wsize = 1000
```

## Distribution Plots

Let's examine data sets that are generated using multivariate normal, gamma, beta, and exponential distributions. 

https://en.wikipedia.org/wiki/Central_limit_theorem 

```python
filename = f'{bfname}_normal.csv'
df = pd.read_csv(f'{folder}/{filename}')
sns.displot(df, x="V1", palette='colorblind')
```

```python
filename = f'{bfname}_gamma.csv'
df = pd.read_csv(f'{folder}/{filename}')
sns.displot(df, x="V1", palette='colorblind')
```

```python
filename = f'{bfname}_beta.csv'
df = pd.read_csv(f'{folder}/{filename}')
sns.displot(df, x="V1", palette='colorblind')
```

```python
filename = f'{bfname}_exp.csv'
df = pd.read_csv(f'{folder}/{filename}')
sns.displot(df, x="V1", palette='colorblind')
```

```python
filename = f'{bfname}_mixed.csv'
df = pd.read_csv(f'{folder}/{filename}')
data = df.iloc[0:2000]
data['dist'] = 'normal'
new_df = df.iloc[2000:5000]
new_df['dist'] = 'gamma'
data = data.append(new_df, ignore_index=True)
new_df = df.iloc[5000:8000]
new_df['dist'] = 'beta'
data = data.append(new_df, ignore_index=True)
new_df = df.iloc[8000:11000]
new_df['dist'] = 'exp'
data = data.append(new_df, ignore_index=True)
sns.displot(data, x="V1", hue="dist", kind="kde",  palette='colorblind')
```
